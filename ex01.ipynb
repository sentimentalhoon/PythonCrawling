```python
url = 'https://search.naver.com/search.naver?where=news&query=%EC%B9%BC%EB%B6%80%EB%A6%BC'
headers = {
    'Accept' : 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7',
    'Accept-Encoding' : 'gzip, deflate, br',
    'Accept-Language' : 'ko-KR,ko;q=0.9,en-US;q=0.8,en;q=0.7,ja;q=0.6',
    'host' : 'www.naver.com',
    'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36'
}
# 페이지 호출
res = requests.get(url, headers = headers)
soup = bs(res.content.decode('utf-8', 'replace'), 'lxml')

news_title = soup.select('a.news_tit')
```

```python
# 순수한 텍스트(컨텐츠) 정보만 저장

text_list = []

for s in news_title:
    text_list.append(s.text)
    
text_list    
```

```python
url = 'https://finance.naver.com/'
headers = {
    'Accept' : 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7',
    'Accept-Encoding' : 'gzip, deflate, br',
    'Accept-Language' : 'ko-KR,ko;q=0.9,en-US;q=0.8,en;q=0.7,ja;q=0.6',
    'host' : 'finance.naver.com',
    'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36'
}
# 페이지 호출
res = requests.get(url, headers = headers)
soup = bs(res.content.decode('euc-kr', 'replace'), 'lxml')
topList = []

seletList = soup.select('#_topItems1 > tr > th > a') 
for value in seletList:        
    topList.append(value.text)

for s in topList:
    print(s)
```
